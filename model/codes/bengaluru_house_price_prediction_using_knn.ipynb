```python
def is_float(x):
  try:
    float(x)
  except:
    return False
  return True
```

```python
df.loc[~df['total_sqft'].apply(is_float)]
```

```python
def convert_to_sqft(x):
  tokens= str(x).split('-')
  if len(tokens) == 2:
    return (float(tokens[0]) + float(tokens[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
df['total_sqft']= df['total_sqft'].apply(convert_to_sqft)
```
